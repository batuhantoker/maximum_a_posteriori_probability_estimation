```python
import numpy as np
from matplotlib import pyplot as plt

def aPosteriori(theta):
    return np.power(theta,alpha_h) * np.power((1-theta),alpha_t) * np.random.beta(3,3)

def plotaPosteriori(theta):
    x = theta
    #alpha_h=alphah
    #alpha_t=alphat
    N=alpha_h+alpha_t
    y = np.array(aPosteriori(x))
    mpa=np.argmax((np.array(aPosteriori(x))))
    plt.title("A Posteriori estimation for N={}".format(N)) 
    plt.xlabel(r'$\hat{\theta}$') 
    plt.ylabel("Estimation") 
    plt.plot(x,y) 
    plt.plot([x[mpa] ,x[mpa]], [0, y[mpa]])
    plt.savefig("figure_a.eps", format="eps", dpi=1000)
    plt.show()

    theta_hat=np.array(np.arange(0,1.01,0.01),dtype=float)
```

```python
alpha_h=6
alpha_t=4
plotaPosteriori(theta_hat)
```
